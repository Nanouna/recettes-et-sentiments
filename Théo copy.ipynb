```python
import pandas as pd
import seaborn as sns
from sklearn.preprocessing import RobustScaler
import matplotlib.pyplot as plt
!pip install fastparquet
```

```python
data = pd.read_parquet('../../data/basic_preproc_recipes_with_reviews_score.parquet')
```

```python
from sklearn.feature_extraction.text import CountVectorizer
count_vectorizer = CountVectorizer( min_df=2, lowercase=True)
bag_of_words = count_vectorizer.fit_transform(data['merged_steps'])
type(bag_of_words)
```

```python
bag_of_words = pd.DataFrame(bag_of_words.toarray(),
columns = count_vectorizer.get_feature_names_out())
```

```python
bag_of_words.set_index(data.index, inplace=True)
```

```python
bag_of_words.head()
```

```python
vectorized_data_wo_bof = data.drop(['tags','description','description', 'merged_steps', 'submitted', 'name', 'merged_ingredients'], axis=1)
vectorized_data_wo_bof
```

```python
bag_of_words['mean_rating'] = vectorized_data_wo_bof['mean_rating']
bag_of_words['review_count'] = vectorized_data_wo_bof['review_count']
```

```python
bag_of_words=bag_of_words.loc[bag_of_words['review_count']>9]
```

```python
bag_of_words
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import cross_validate
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
cv_nb = cross_validate(
    model,
    bag_of_words.drop(columns='mean_rating'),
    bag_of_words['mean_rating'],
    cv=5,
    n_jobs=1,
    scoring='r2'
)

#round(cv_nb['test_score'].mean(),2)
cv_nb
```
